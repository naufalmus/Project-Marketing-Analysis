# Customer Segmentation

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import plotly.express as px

# 1. Pilih kolom untuk segmentasi (RFM / Demografi Keuangan)
# Asumsi nama dataframe utama Anda adalah 'df'
features = ['Income', 'Total_Spending', 'Recency']
df_cluster = df[features].copy()

# 2. Tangani Missing Values (K-Means tidak bisa memproses data kosong/NaN)
df_cluster = df_cluster.dropna()

# 3. Standardisasi Data (SANGAT PENTING untuk K-Means)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_cluster)

# Mengubah hasil array kembali menjadi dataframe untuk memudahkan
df_scaled = pd.DataFrame(df_scaled, columns=features)
```

```python
# Menghitung WCSS (Within-Cluster Sum of Square) atau Inertia
wcss = []
K_range = range(1, 11) # Mencoba dari 1 hingga 10 cluster

for k in K_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

# Visualisasi Elbow Method
plt.figure(figsize=(10, 6))
plt.plot(K_range, wcss, marker='o', linestyle='--', color='b')
plt.title('Elbow Method Untuk Menentukan Jumlah Cluster Optimal', fontsize=14)
plt.xlabel('Jumlah Cluster (K)')
plt.ylabel('WCSS (Inertia)')
plt.xticks(K_range)
plt.grid(True)
plt.show()
```

```python
# Tentukan jumlah cluster (Ubah angka 3 sesuai hasil Elbow Method Anda)
optimal_k = 3

# Buat dan latih model
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42)
clusters = kmeans.fit_predict(df_scaled)

# Masukkan label cluster kembali ke Dataframe asli (yang sudah di dropna)
# Gunakan .loc agar indexnya tetap sejajar
df_hasil = df.loc[df_cluster.index].copy()
df_hasil['Customer_Segment'] = clusters

# Ubah nama label cluster agar lebih rapi (opsional)
df_hasil['Customer_Segment'] = df_hasil['Customer_Segment'].map({
    0: 'High-Value Customer',     # Sesuaikan index ini dengan hasil tabel Anda nanti
    1: 'At Risk Customer', 
    2: 'Mid Value Customer'
})

# Cek rata-rata pengeluaran, pendapatan, dll di setiap cluster
cluster_profile = df_hasil.groupby('Customer_Segment')[features].mean().round(2)
display(cluster_profile)
```

```python
plt.figure(figsize=(12, 8))

# Scatter plot Income vs Total Spending, diwarnai berdasarkan hasil Cluster
sns.scatterplot(
    data=df_hasil, 
    x='Income', 
    y='Total_Spending', 
    hue='Customer_Segment', 
    palette='Set1', 
    s=100, # Ukuran titik
    alpha=0.7 # Transparansi agar titik yang bertumpuk terlihat
)

plt.title('Customer Segmentation: Income vs Total Spending', fontsize=16, fontweight='bold')
plt.xlabel('Income', fontsize=12)
plt.ylabel('Total Spending', fontsize=12)
plt.legend(title='Segment', bbox_to_anchor=(1.05, 1), loc='upper left')

# Agar angka di sumbu X dan Y lebih mudah dibaca (tidak menggunakan scientific notation)
plt.ticklabel_format(style='plain', axis='both') 
plt.show()
```
